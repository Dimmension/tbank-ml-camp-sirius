```python
import json
```

```python
with open('../data/data_full_spoiled.json', 'r') as f:
    data = json.load(f)
```

```python
labels_with_examples = {} 
for e in data['val']:
    labels_with_examples[e[1]] = []
```

```python
for e in data['val']:
    labels_with_examples[e[1]].append(e[0])
```

```python
for label, descriptions in labels_with_examples.items():
    filtered_descriptions = [
        descriptions[0],
    ]
    print(filtered_descriptions)
    labels_with_examples[label] = filtered_descriptions
```

```python
labels = []
for label, descriptions in labels_with_examples.items():
    labels.append(f'\n{label}\n')
    for description in descriptions:
        labels.append(f'{description}\n')
```

```python
labels
```

```python
with open('../data/labels.txt', 'w') as f:
    f.writelines(labels)
```

```python
data = {}
with open('../data/labels_with_description.txt', "r") as file:
    for line in file:
        if ": " in line:
            key, value = line.split(": ", 1)
            data[key.strip()] = value.strip()

with open('../data/labels_with_description.json', "w") as file:
    json.dump(data, file, indent=4)
```
